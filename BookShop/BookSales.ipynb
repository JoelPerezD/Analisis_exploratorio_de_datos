```python
plt.hist(df['Publishing Year'] , bins=20, color='blue', alpha=0.7)
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Distribution of Publishing Year') 
plt.show()
```

```python
df['genre'].value_counts().plot(kind='bar', color='blue', alpha=0.7)
plt.xlabel('Genero')
plt.ylabel('Numero de libros')
plt.title('Numero de libros por genero')
plt.show()
#cuantos generos hay en el dataset, kind=bar para que se vea como un histograma, color y al apha para el color y la transparencia
```

```python
#agrupamos por autor y contamos el numero de libros por autor
df.groupby('Author')['Book_average_rating'].mean().head(10)
#nos indica el promedio de rating por autor, y nos muestra los 10 primeros autores con el rating mas alto
```

```python
df.groupby('Author')['gross sales'].sum().sort_values(ascending=False).head(10)
# Nos indica cuales son los autores con la mayor cantidad de ventas 
```

```python
sns.boxplot(x='genre', y='Book_ratings_count', data=df)
plt.xlabel('Genero')
plt.ylabel('Numero de ratings')
plt.title('Numero de ratings por genero')
plt.show()
#boxplot para ver la distribucion de los ratings por genero, nos indica que el genero de ficcion tiene la mayor cantidad de ratings, y el de no ficcion la menor cantidad de ratings
#vemos la distribucion de los ratings por genero, y vemos que el genero de ficcion tiene la mayor cantidad de ratings, y el de no ficcion la menor cantidad de ratings
```

```python
# con scatter podemos identificar la relacion enrtre ambas variables, y ver si hay alguna correlacion entre ellas
plt.scatter(df['sale price'], df['units sold'])
plt.xlabel('Precio de venta')
plt.ylabel('Unidades vendidas') 
plt.title('Precio de venta vs Unidades vendidas')
plt.show()
```

Podemos obervar hay mas unidades vendidas a menor precio de venta, por lo que podria ser un indicador de de que existe una correlacion entre ambas variables 

```python
lenguage_counts = df['language_code'].value_counts()
plt.pie(lenguage_counts, labels=lenguage_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribucion de libros por idioma')
#plt pie se ingresa el conteo, labels es para las etiquetas, autopct es para el porcentaje, y startangle es para el angulo de inicio del grafico
plt.axis('equal')  # equeal es para que el grafico sea un circulo
plt.show()
```

```python
df.groupby('Publisher ')['publisher revenue'].mean().sort_values(ascending=False).head(10)
# agrupamos por editorial y vemos el promedio de ingresos por editorial, y los 10 primeros con mayor ingreso
#vemos que la editorial con mayor ingreso es la de Random House, y la que menos ingresos tiene es la de HarperCollins
```

```python
df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending=False).head(10)
#agrupamos por editorial y vemos el total de ingresos por editorial, y los 10 primeros con mayor ingreso
```

```python
df.groupby('Author_Rating')['Book_ratings_count'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='blue', alpha=0.7)
plt.xlabel('Rating del autor')
plt.ylabel('Numero de ratings')
plt.title('Numero de ratings por rating del autor')

plt.show()
```

```python
plt.scatter(df['Book_average_rating'], df['Book_ratings_count'], alpha=0.5)     
plt.xlabel('Rating del libro')
plt.ylabel('Numero de ratings')
plt.title('Rating del libro vs Numero de ratings')
plt.show()
```

```python
total_sales_by_author = df.groupby('Author')['gross sales'].sum().sort_values(ascending=False).head(10)
total_sales_by_author 
```

```python
plt.figure(figsize=(10, 6))
plt.bar(total_sales_by_author.index, total_sales_by_author.values, color='blue', alpha=0.7)
```

```python
# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['float64', 'int64'])
#select dtypers, inculye flotantes e ints
#elecctionr unicamente valores numericos
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix') 
plt.show()
# el annot es para que se vean los valores de la correlacion, cmap es el color del mapa, y el fmt es para el formato de los numeros
```
