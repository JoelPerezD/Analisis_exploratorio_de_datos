```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#eliminamos los valres duolicaods 
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
sns.countplot(data=df, x='Sex', hue='Medal')
plt.title('Distribución de medallas por sexo')
plt.show()  
```

```python
sns.histplot(data=df, x='Age', bins=20, kde=True, color='blue', alpha=0.5)
plt.title('Distribución de edades')
plt.xlabel('Edad')  
plt.ylabel('Distribución')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(data=df, x='Height', bins=20, kde=True, color='green', alpha=0.5, ax=ax[0])
ax[0].set_title('Distribución de alturas')
ax[0].set_xlabel('Altura')
ax[0].set_ylabel('Distribución')  
sns.histplot(data=df, x='Weight', bins=20, kde=True, color='red', alpha=0.5, ax=ax[1])
ax[1].set_title('Distribución de alturas')
ax[1].set_xlabel('Altura') 
plt.ylabel('Distribución')
plt.show()
# Correlación entre variables numéricas
```

```python
# hasta el momento hemos visto que los hombres son los que más medallas ganan, pero no hemos visto si hay alguna correlación entre las variables numéricas. Para ello, vamos a usar el método corr() de pandas y la función heatmap() de seaborn para visualizar la matriz de correlación.
# seleccionamos solo las columnas numéricas
variables_numericas = df.select_dtypes(include=[np.number])
correlation_matrix = variables_numericas.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlación')
plt.show()
```

podemos ver que hay correlaciones obvias entre la estatura y el peso dentro de las variables numericas, pero no estan altamente relacionadas. 

```python
# desde que observamos que los hombres son los que mas ganasn medallas, contamos cuantos hombres y mujeres hay en el dataset    
df['Sex'].value_counts().plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribución de hombres y mujeres')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.show()
```

```python
#hay mucho mas hombre que mujeres,  por lo que tiene sentido que haya mas hombres ganando medallas
# pero no sabemos si esto es un problema de la muestra o si en realidad hay mas hombres
#  que mujeres en los juegos olimpicos. Para ello, 
# vamos a ver la cantidad de hombres y mujeres que hay en cada deporte.
```

```python
# 'Sport' column does not exist. Let's use 'City' as an example.
df['Sport'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title('Top 10 deportes más populares')
plt.xlabel('deporte')
plt.ylabel('participantes')
plt.show()
```

```python
#deportes por sexo
df.groupby(['Sex', 'Sport']).size().sort_values().head(10).plot(kind='barh', color=['blue', 'pink'])
plt.title('Distribución de deportes por sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.show()
```

```python
conteo = df.groupby(['Sport', 'Sex']).size().unstack().fillna(0)

# Tomamos los 10 deportes con más participantes en total
top_deportes = conteo.sum(axis=1).sort_values(ascending=False).head(10).index
conteo_top = conteo.loc[top_deportes]

# Gráfico de barras agrupadas
conteo_top.plot(kind='bar', figsize=(10,6), color=['orange', 'skyblue'])
plt.title('Cantidad por Sexo en los 10 Deportes más Populares')
plt.xlabel('Deporte')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.legend(title='Sexo')
plt.tight_layout()
plt.show()
```

```python
#distribuciion de las medallas por año
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Year', hue='Medal')
plt.title('Distribución de medallas por año')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

Observamos que se han incrementado el numero de medallas a lo largo de los anios

```python
year_avg_age= df.groupby('Year')['Age'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=year_avg_age, x='Year', y='Age', marker='o')
plt.title('Edad promedio de los atletas por año')
plt.xlabel('Año')
plt.ylabel('Edad promedio')
plt.xticks(rotation=45)
plt.show()
```

Obsevamos que la edad promedio los atletas era mayor a inicios del siglo pasado

```python
# Obtenemos el pais que mas medallas de oro ha ganado
paies_medallas_oro = df[df['Medal'] == 'Gold']['NOC'].value_counts().head(10)
plt.figure(figsize=(12, 6))
paies_medallas_oro.plot(kind='bar', color='gold')
plt.title('Top 10 paises con más medallas de oro') 
plt.xlabel('País')
plt.ylabel('Cantidad de medallas de oro')
plt.xticks(rotation=45)
plt.show()
```

```python
# Contar la cantidad de eventos únicos por deporte
sports_event_count = df.groupby('Sport')['Event'].nunique().sort_values(ascending=False)
sports_event_count.plot(kind='bar', figsize=(12, 6), color='purple')
plt.title('Cantidad de eventos por deporte')
plt.xlabel('Deporte')
plt.ylabel('Cantidad de eventos')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.violinplot(data=df, x='Medal', y='Height')
plt.title('Distribución de altura por medalla')
plt.xlabel('Medalla')
plt.ylabel('Altura')
plt.xticks(rotation=45)
plt.show()
# Gráfico de dispersión de altura vs peso
```
